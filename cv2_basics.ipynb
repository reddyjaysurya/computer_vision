 ### Blending images


```python
img1 = cv2.cvtColor(cv2.imread("DATA/overlap_coins.jpg"),cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.cvtColor(cv2.imread("DATA/watermark_no_copy.png"), cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1 = cv2.resize(img1,(1280,1277))
img2 = cv2.resize(img2,(1280,1277))
```

```python
plt.imshow(img2)
```

### add weighted function only works with images of same size

```python
img3 = cv2.addWeighted(src1 = img1, alpha = 0.5, src2 = img2,beta = 0.05,gamma = 0)
```

```python
plt.imshow(img3)
```
